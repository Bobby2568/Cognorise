```python
df.drop(['name'],axis=1,inplace=True)
```

```python
df.head()
```

```python
x=df.drop(['rating','mfr','shelf','type','carbo','vitamins','weight','cups'],axis=1)
y=df['rating']
```

```python
# splitting dataset as train and test set

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x_train.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
model=LinearRegression()
model.fit(x_train,y_train)
print(cross_val_score(model,x,y,cv=10).mean())
print(model.score(x,y))
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import accuracy_score,r2_score,mean_squared_error
from sklearn import tree
from math import sqrt

tree=DecisionTreeRegressor()
tree.fit(x_train,y_train)
y_pred_train=tree.predict(x_train)
y_pred=tree.predict(x_test)
print(sqrt(mean_squared_error(y_test,y_pred)))
```

```python
from sklearn.model_selection import GridSearchCV
params={
    'max_depth':[10,15,20,40,50],
    'min_samples_split':[2,4,6,8,10],
    'min_samples_leaf':[2,4,5,7,8]}
model=GridSearchCV(tree,params,n_jobs=-1)
model.fit(x_train,y_train)
print(model.best_params_)
```

```python
tree1=DecisionTreeRegressor(max_depth=40,min_samples_leaf=2,min_samples_split=2)
tree1.fit(x_train,y_train)
y_pred=tree1.predict(x_test)
print(r2_score(y_test,y_pred))
print(sqrt(mean_squared_error(y_test,y_pred)))
```
