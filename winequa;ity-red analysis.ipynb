```python
# drop target values from the x variable

x=df[['volatile acidity','citric acid','pH','sulphates']].values
```

```python
y=df[['alcohol']].values
```

```python
x=sc.fit_transform(x)
```

```python
from sklearn.ensemble import AdaBoostRegressor,RandomForestRegressor,GradientBoostingRegressor
from sklearn.linear_model import LinearRegression,Ridge
from sklearn.tree import DecisionTreeRegressor
```

```python
models={
    'LinearRegression': LinearRegression(),
    'Ridge':Ridge(),
    'AdaBoostRegressor':AdaBoostRegressor(),
    'DecisionTreeRegressor': DecisionTreeRegressor(),
    'RandomForestRegressor': RandomForestRegressor(),
    'GradientBoostingRegressor': GradientBoostingRegressor()
}
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from sklearn.model_selection import cross_val_score
```

```python
for name, model in models.items():
    scores = cross_val_score(model,x,y,scoring='neg_mean_squared_error',cv=20)
    print('Cross validation model:{}'.format(name))
    rmse = np.sqrt(-scores)
    rmse_average = np.mean(rmse)
    print('Average RMSE:',rmse_average)
    print('*'*100)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
model=RandomForestRegressor()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x)
```

```python
output= pd.DataFrame(zip(y,y_pred),columns=('Actual','Prediction'),dtype=float)
```
